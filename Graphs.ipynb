```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [10,12]
y = [2,8]
plt.plot(x, y, marker='*', color = 'blue', linestyle='-', linewidth = 5, markersize = 15)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
plt.scatter(x, y, color = 'red', marker='x')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
Name = ['Poonam','Dimple']
Age = [20,30]
plt.bar(Name, Age, color = 'green')
plt.title('Bar plot')
plt.xlabel('Name')
plt.ylabel('Age')
plt.grid()
plt.show()
```

```python
data = [2,3,4,7,8,9]
plt.hist(data, bins=10, color='lightpink', edgecolor ='black')
plt.title('Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

```python
sizes = [20,48,38,79]
labels = ['a','b','c','d']
plt.pie(sizes, labels=labels, autopct = '%1.1f%%', startangle = 0)
plt.title('Pie Chart')
plt.show()
```
