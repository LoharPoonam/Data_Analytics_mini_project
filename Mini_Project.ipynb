```python
# Filter dataset where Distance_km is between 5 and 15 and Route is "Los Angeles"
filtered_df = df[(df["Distance_km"] >= 5) & (df["Distance_km"] <= 15) & (df["Route"] == "Los Angeles")]

# Group by Delivery_Status and calculate mean & median of Delivery_Time
aggregated_df = filtered_df.groupby("Delivery_Status")["Delivery_Time"].agg(["mean", "median"]).reset_index()

# Sort by mean Delivery_Time in descending order
aggregated_df = aggregated_df.sort_values(by="mean", ascending=False)

print(aggregated_df)
```

```python
# Ensure the column is of string type
df["Full_Name"] = df["Customer_Name"].astype(str)

# Handle cases where there may not be a last name (avoid NaN issues)
df[["First_Name", "Last_Name"]] = df["Full_Name"].str.split(n=1, expand=True).fillna("")

# Group by Route and count unique Full_Name entries
unique_customers = df.groupby("Route")["Full_Name"].nunique().reset_index()
unique_customers.rename(columns={"Full_Name": "Unique_Customers"}, inplace=True)

print(unique_customers)
```

```python
# Normalize the Price column between 0 and 1 manually
df["Normalized_Price"] = (df["Price"] - df["Price"].min()) / (df["Price"].max() - df["Price"].min())

print(df[["Price", "Normalized_Price"]].head())  # Check the normalized values


# Replace negative Delivery_Time values with the median
median_delivery_time = df[df["Delivery_Time"] >= 0]["Delivery_Time"].median()
df["Delivery_Time"] = df["Delivery_Time"].apply(lambda x: median_delivery_time if x < 0 else x)

# Sample product_data DataFrame
product_data = pd.DataFrame({
    "Order_ID": df["Order_ID"].sample(10),  # Taking 10 random Order_IDs from df
    "Product_Name": ["Product_" + str(i) for i in range(10)],
    "Category": ["Category_" + str(i % 3) for i in range(10)]
})

# Merge with product_data on Order_ID
merged_df = pd.merge(df, product_data, on="Order_ID", how="left")

# Display first 5 rows of merged dataset
print(merged_df.head())
```
